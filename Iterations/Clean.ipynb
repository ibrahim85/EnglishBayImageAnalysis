```python
from scipy.misc import imread
import pandas as pd
import numpy as np
import glob
import re
```

```python
# flatten in imread set to true, images loaded as gray scale
row = 0
image_data = pd.DataFrame(columns=("image", "timestamp"))

for path in sorted(glob.glob("./katkam-scaled/*")):
    
    matches = re.search("\d+", path)
    image_data.loc[row] = [imread(path, flatten=True).flatten(), matches.group()]
    
    #image_data.loc[row] = [imread(path, flatten=True).flatten(), matches.group()]
    row += 1

image_data["timestamp"] = pd.to_datetime(image_data["timestamp"])
```

```python
# get list of weather data csv files
files = [f for f in glob.glob("./yvr-weather/*")]
```

```python
filename = "./yvr-weather/weather-51442-201606.csv"
weather_data = pd.read_csv(filename, sep=',', skiprows=16, parse_dates=[0])
```

```python
# data quality <-- look at later? 
weather_data.drop(["Year","Month","Day","Time","Temp Flag", "Dew Point Temp Flag",
                   "Rel Hum Flag", "Wind Dir Flag", "Wind Spd Flag",
                   "Visibility Flag", "Stn Press Flag",
                   "Hmdx Flag", "Wind Chill Flag", "Hmdx", "Wind Chill",
                   "Data Quality"], axis=1, inplace=True)
```

```python
weather_data
```

```python
test = weather_data[weather_data["Weather"].isnull()]
train = weather_data[~weather_data["Weather"].isnull()]

times = list(train['Date/Time'].map(lambda x: x))
test_times = list(test['Date/Time'].map(lambda x: x))

X_img_train = image_data[image_data.timestamp.isin(times)]
X_img_test = image_data[image_data.timestamp.isin(test_times)]

times = list(X_img_train.timestamp.map(lambda x : x))
test_times = list(X_img_test.timestamp.map(lambda x : x))


y_train = weather_data[weather_data['Date/Time'].isin(times)]
y_test = weather_data[weather_data['Date/Time'].isin(test_times)]
```

```python
# reshape to (153,49152)
images = np.zeros((153,49152), dtype=np.float32)
for i in range(153):
    images[i] = X_img_train.image.iloc[0]
```

```python


from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)
model.fit(images, y_train.Weather)
```
